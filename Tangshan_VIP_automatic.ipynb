```python
from matplotlib.lines import segment_hits
import re
import numpy as np
import sys
import matplotlib.pyplot as plt
import pdb
 
'''
This function accepts a list of filenames and return
a list of file pointers corresponding to those files
NOT USED NOW
'''
 
#find the closest number to K in the list
def closest(lst, K):     
    return lst[min(range(len(lst)), key = lambda i: abs(lst[i]-K))] 
 
#find the start and end line numbers for textgrid from automatic cut
#segment is a boolean indicating whether there is an #END# segment indicator
def line_numbers_textgrid_annotater(file_path: str, segment: bool = False) -> (int,int):
    with open(file_path, "r", encoding='UTF8') as f:
        #get the entire text in a list
        text = f.readlines()
        #store the indices in an array
        indices = []
        #find the line number of the word "phones"
        for i in range(0, len(text)):
            if "phones" in text[i] or "syllables" in text[i]:
                start_line = i
                break
        #truncate all the useless lines
        text = text[start_line + 1:]
        #search for matches for all lines and record line numbers
        for i in range(0, len(text)):
            if re.search(r"^\"[a-zA-Z]+\d\"\n$", text[i]) or (segment and text[i].strip('\n').strip('"') == "#END#"):
                indices.append(i)
        #NOTE: the end index has a +1 because we will feed it into python range() function
        start = indices[0] + start_line + 1; end = indices[-1] + start_line + 1 + 1
    f.close()
    return start,end
 
#find the start and end line numbers for pitchtier
def line_numbers_pitchtier(file_path:str, start_time:int or str, end_time:int or str) -> (int,int):
    with open(file_path, "r", encoding='UTF8') as f:
        #get the entire text in a list
        text = f.readlines()
        #clean out the metadata
        text = text[6:len(text)]
        #convert all entries in the list to float
        text_float = [float(i) for i in text]
        #search for the indices that is closest to the start time and end time
        start = closest(text_float, float(start_time))
        end = closest(text_float, float(end_time))
        for i in range(0, len(text_float)):
            if text_float[i] == start:
                start_idx = i + 6
            if text_float[i] == end:
                end_idx = i + 6 + 1 + 1
    f.close()
    return start_idx,end_idx
 
#function to extract out the pitches in time domain
def get_pitches(file_path: str, start_line:int, end_line: int) -> (list, list):
    time_spots = []
    pitches = []
    with open(file_path, "r") as f:
        lines = f.readlines()
        for line in range(start_line, end_line):
            #time spots should be at even indices and pitches at odd indices(zero indexing!!!!)
            if line % 2 == 0:
                time_spots.append(float(lines[line].strip('\n')))
            else:
                pitches.append(float(lines[line].strip('\n')))
    f.close()
    return time_spots, pitches
 
#function to extract out annotations in TextGrid files coming from the aligner
# segment indicates whether there is an "#END#" marker 
def get_annotations_from_aligner(file_path: str, start_line: int, end_line: int, segment: bool = False) -> list:
    annotations = []
    with open(file_path, "r", encoding='UTF8') as f:
        lines = f.readlines()
        for line in range(start_line,end_line):
            #check whether the line is an annotation
            label = lines[line].strip('\n').strip('"')
            #search the pattern that starts with letters and ends with a digit(韵母)
            if re.search("^[a-zA-Z]+\d$", label) is not None:
                #convert the times to int
                interval = (float(lines[line - 2].strip('\n')), float(lines[line - 1].strip('\n')))
                annotations.append((interval, label))
            if label == "#END#" and segment:
                annotations.append(label)
    f.close()
    return annotations
 
#function to evenly space points inside an interval, and extract the pitches at these points
def sampling(interval: tuple, time_spots: [], pitches: []) -> np.array:
    sampled_pitches = np.array(pitches)
    #find the closest number to the interval lowerbound
    start = closest(time_spots, interval[0])
    start_idx = time_spots.index(start)
    #find the closest number to the interval upperbound
    end = closest(time_spots, interval[1])
    end_idx = time_spots.index(end)
    #evenly space out points in the array
    indices_all = np.linspace(start = start_idx, stop = end_idx, num = end_idx - start_idx + 1, dtype = int)
    indices_10 = np.linspace(start = start_idx, stop = end_idx, num = 10, dtype = int)
    sampled_pitches_10 = sampled_pitches[indices_10]
    sampled_pitches_all = sampled_pitches[indices_all]
    return sampled_pitches_10, sampled_pitches_all
 
#function to find the linear regression of the syllable
def linear_reg(pitches_sing: np.array):
  index = list(range(1,len(pitches_sing) + 1))
  sing_fit = np.polyfit(y=pitches_sing, x=index, deg = 1)
  return sing_fit
 
#function to visualizae fit plots
def visualize(x_coord_s, y_sing, y_sing_fit, syl, data_type):
  print("Visualization of the {vtype} fit graph for syllable {name}".format(name=syl, vtype=data_type))
  if data_type != "Cross-Correlation Distribution":
    ax1 = plt.plot(x_coord_s, y_sing, '.', x_coord_s, y_sing_fit, '-')
    plt.show()

  return None
 
#function to clean the array by dropping out values <= Q1 - 1.5IQR or >= Q3 + 1.5 IQR
def drop_outliers(sampled_pitches_sing: np.array) -> (np.array):
  q3_sing, q1_sing = np.percentile(sampled_pitches_sing, [75,25])
  iqr_sing = q3_sing - q1_sing
  cleaned_pitches_sing = np.array([pitch for pitch in sampled_pitches_sing 
                                   if pitch >= q1_sing - 1.5 * iqr_sing and pitch <= q3_sing + 1.5 * iqr_sing])
  return cleaned_pitches_sing
 
#function to set up the visualization parameters
def setup_params(pitches_sing, syl):
  #set up the linear regression
  linear_fit = linear_reg(pitches_sing)
  #polynomial fit of the sing version
  poly_sing = np.poly1d(linear_fit)
  print("linear fit of {syllable} is y = {slf}".format(syllable=syl, slf=poly_sing))
 
  ##set up the visualization parameters of the fit
  #set x coordinates of sing pitches fit
  x_coord_sing = list(range(1,len(pitches_sing) + 1))
  y_sing_fit = poly_sing(x_coord_sing)
  y_sing = pitches_sing

  #return the parameters
  return y_sing, x_coord_sing, y_sing_fit

def print_result(s, r, n):
  if n == 0:
    p = 0
  else:
    p = len(r)/n
  print("Percentage of real {s1} tones in the song: {p1}, r:{r1},n:{n1}".format(s1=s, p1=p, r1=len(r), n1=n))
  return 

 
'''
HOW TO USE THIS SCRIPT:
1. Copy the file paths of your TextGrid file and PitchTier file from Google Colab File Explorer (the left window pane)
and paste them to the corresponding variables.
2. Change the variable "song_name" to the song you are investigating. For example, if the song is "橄榄树", you should 
change it to "橄榄树" or "ganlanshu". 
3. Click the Run button in Google Colab. You can then see the RMSE output in the console. 
4. That is all you need to do! 😀
'''

def main():
    # TODO: Change the file path to that of the song you are calculating
    sing_textgrid_path = "/content/piaoyi.TextGrid"
    sing_pitchtier_path = "/content/piaoyi.PitchTier"

    #TODO:  Change the song name to that of the song you are analyzing
    song_name = "漂移20s"
    # Software segmentation mode or manual praat mode

    #extract out annotated points and pitches in the song version
    start_sing,end_sing = line_numbers_textgrid_annotater(sing_textgrid_path, segment=True)
    #print(start_sing,end_sing)
    sing_annotations = get_annotations_from_aligner(sing_textgrid_path, start_sing, end_sing, segment=True)
  
    selected_tones = list()
    for idx in range(0,len(sing_annotations)):
      if sing_annotations[idx] != "#END#":
        if int(sing_annotations[idx][1][-1]) in [0,1,2,3,4,5]:
          selected_tones.append(idx)
 
    start_sing,end_sing = line_numbers_pitchtier(sing_pitchtier_path,sing_annotations[0][0][0],sing_annotations[-2][0][1])
    sing_time_spots, sing_pitches = get_pitches(sing_pitchtier_path, start_sing, end_sing)

    # the segment array
    segment = []
    # the average of each segment
    segment_avg = 0.0
    # the tone-pitch_avg dict
    tone_pitchAvg = {}

    # number of tones that is LABELED AS first tone
    num_first_tones_labeled = 0
    # number of tones that is LABELED AS second tone
    num_second_tones_labeled = 0
    # number of tones that is LABELED AS third tone
    num_third_tones_labeled = 0
    # number of tones that is LABELED AS fourth tone
    num_fourth_tones_labeled = 0
    # number of tones that is LABELED AS neutral tone
    num_neutral_tones_labeled = 0
    # list of invalid results (slope >10/<-10 for 2nd/4th tones or <5 data points)

    # list of real tones
    real_first_tones = []
    real_third_tones = []
    real_second_tones = []
    real_fourth_tones = []
    real_neutral_tones = []

    # number of ending tones that is LABELED AS ending first tone
    num_ending_first_tones_labeled = 0
    # number of ending tones that is LABELED AS ending second tone
    num_ending_second_tones_labeled = 0
    # number of ending tones that is LABELED AS ending third tone
    num_ending_third_tones_labeled = 0
    # number of ending tones that is LABELED AS ending fourth tone
    num_ending_fourth_tones_labeled = 0
    # number of ending tones that is LABELED AS ending neutral tone
    num_ending_neutral_tones_labeled = 0

    # list of real ending tones
    real_ending_first_tones = []
    real_ending_third_tones = []
    real_ending_second_tones = []
    real_ending_fourth_tones = []
    real_ending_neutral_tones = []

    # list of invalid results
    invalid_results = []

    print("**********{song_name}**********".format(song_name = song_name))

    for i in range(0, sing_annotations.__len__()):

        #ending syllable marker
        end = False
        #set marker to indicate ending syllable
        try:
          if sing_annotations[i+1] == "#END#":
            end = True
        except:
          pass

        # if the annotation is "#END#", calculate the average of the segment and skip the following lines
        if sing_annotations[i] == "#END#":
          print("Reach the end of segment")
          segment_avg = np.mean(segment)
          print("This segment average is: {seg_avg}".format(seg_avg=segment_avg))
          for i, key in enumerate(tone_pitchAvg.keys()):

            # set marker to indicate ending syllable
            if i == len(tone_pitchAvg.keys()) - 1:
              end = True

            # if the first tone pitch average is higher than the segment average, it's considered as a first tone
            if key[-1] == '1':
              num_first_tones_labeled += 1
              if end:
                num_ending_first_tones_labeled += 1
              if tone_pitchAvg[key] >= segment_avg:
                print("Syllable {syl} is considered as a first tone because its pitch average is greater than the segment average.".format(syl=key))
                real_first_tones.append(key)
                if end:
                  real_ending_first_tones.append(key)
            # if the third tone pitch average is lower than the segment average, it's considered as a third tone
            elif key[-1] == '3':
              num_third_tones_labeled += 1
              if end:
                num_ending_third_tones_labeled += 1
              if tone_pitchAvg[key] < segment_avg:
                print("Syllable {syl} is considered as a third tone because its pitch average is less than the segment average.".format(syl=key))
                real_third_tones.append(key)
                if end:
                  real_ending_third_tones.append(key)

          segment.clear()
          tone_pitchAvg.clear()
          continue

        #use original data
        sampled_pitches_sing, sampled_pitches_sing_original  = sampling(sing_annotations[i][0], sing_time_spots, sing_pitches)
        # skip syllables with less than 5 data points
        if len(sampled_pitches_sing_original) < 5:
          print("{syl} has less than 5 data points. Adding to invalid results...".format(syl=sing_annotations[i][1]))
          invalid_results.append(sing_annotations[i][1])
          continue

        # copy the pitch values to the segment array
        segment.extend(sampled_pitches_sing_original)
        # calculate the pitch averages
        print("calculating the mean pitch for each tone")
        pitch_avg = np.mean(sampled_pitches_sing_original)
        if sing_annotations[i][1] not in tone_pitchAvg.keys():
          tone_pitchAvg[sing_annotations[i][1]] = pitch_avg
        else:
          x = "1"
          while x + sing_annotations[i][1] in tone_pitchAvg.keys():
            x = int(x) + 1
            x = str(x)
          tone_pitchAvg[x + sing_annotations[i][1]] = pitch_avg

        #use cleaned data for 2nd/4th tone slope calculations
        cleaned_pitches_sing_orig = drop_outliers(sampled_pitches_sing_original)

        if sing_annotations[i][1][-1] == '2':
          slope = linear_reg(cleaned_pitches_sing_orig)
          slope = slope[0]
          if slope < 0 and slope > -10:
              real_second_tones.append(sing_annotations[i][1])
              if end:
                real_ending_second_tones.append(sing_annotations[i][1])
          if slope < 10 and slope > -10:
              num_second_tones_labeled += 1
              if end:
                num_ending_second_tones_labeled += 1
          else:
              print("Slope too large. Adding {syl} to invalid results...".format(syl=sing_annotations[i][1]))
              invalid_results.append(sing_annotations[i][1])

        if sing_annotations[i][1][-1] == '4':
          slope = linear_reg(cleaned_pitches_sing_orig)
          slope = slope[0]
          if slope < 0 and slope > -10:
              real_fourth_tones.append(sing_annotations[i][1])
              if end:
                real_ending_fourth_tones.append(sing_annotations[i][1])
          if slope < 10 and slope > -10:
              num_fourth_tones_labeled += 1
              if end:
                num_ending_fourth_tones_labeled += 1
          else:
              print("Slope too large. Adding {syl} to invalid results...".format(syl=sing_annotations[i][1]))
              invalid_results.append(sing_annotations[i][1])

        if i in selected_tones:

                #build up visualization for the uncleaned data

                y_sing, x_coord_sing, y_sing_fit = setup_params(sampled_pitches_sing_original, sing_annotations[i][1])
                
                visualize(x_coord_sing,y_sing, y_sing_fit, sing_annotations[i][1], "pitches_orig")
                
                #build up visualization for the cleaned data

                y_sing, x_coord_sing, y_sing_fit = setup_params(cleaned_pitches_sing_orig, sing_annotations[i][1])

                visualize(x_coord_sing,y_sing, y_sing_fit, sing_annotations[i][1], "pitches_cleaned")

    print()
    print("TANGSHAN DIALECT")
    print()
    print("Labeled first tones that are considered as real first tones based on average pitches: {tone_list}".format(tone_list=real_first_tones))
    print_result( "first",real_first_tones,num_first_tones_labeled )
    print("Labeled second tones that are considered as real second tones based on slope: {tone_list}".format(tone_list=real_second_tones))
    print_result( "second", real_second_tones,  num_second_tones_labeled )
    print("Labeled third tones that are considered as real third tones based on average pitches: {tone_list}".format(tone_list=real_third_tones))
    print_result( "third", real_third_tones,  num_third_tones_labeled )
    print("Labeled fourth tones that are considered as real fourth tones based on slope: {tone_list}".format(tone_list=real_fourth_tones))
    print_result( "fourth", real_fourth_tones,  num_fourth_tones_labeled )  
    print("Labeled neutral tones that are considered as real neutral tones based on average pitch of preceding syllable: {tone_list}".format(tone_list=real_neutral_tones))
    print_result( "neutral", real_neutral_tones,  num_neutral_tones_labeled )
    print()
    sum_r = len(real_first_tones) + len(real_third_tones) + len(real_second_tones) + len(real_fourth_tones) +len(real_neutral_tones)
    sum_n = num_first_tones_labeled + num_third_tones_labeled + num_second_tones_labeled + num_fourth_tones_labeled + num_neutral_tones_labeled
    print( "Congruence index percentage: {pct}, r:{r}, n:{n}".format(pct=sum_r/sum_n,r=sum_r,n=sum_n))
    print()
    print("Labeled ENDING first tones that are considered as real ENDING first tones based on average pitches: {tone_list}".format(tone_list=real_ending_first_tones))
    print_result( "first", real_ending_first_tones,  num_ending_first_tones_labeled )
    print("Labeled ENDING second tones that are considered as real ENDING second tones based on slope: {tone_list}".format(tone_list=real_ending_second_tones))
    print_result( "second", real_ending_second_tones,  num_ending_second_tones_labeled )
    print("Labeled ENDING third tones that are considered as real ENDING third tones based on average pitches: {tone_list}".format(tone_list=real_ending_third_tones))
    print_result( "third", real_ending_third_tones,  num_ending_third_tones_labeled )
    print("Labeled ENDING fourth tones that are considered as real ENDING fourth tones based on slope: {tone_list}".format(tone_list=real_ending_fourth_tones))
    print_result( "fourth", real_ending_fourth_tones,  num_ending_fourth_tones_labeled )  
    print("Labeled ENDING neutral tones that are considered as real ENDING neutral tones based on average pitch of preceding syllable: {tone_list}".format(tone_list=real_ending_neutral_tones))
    print_result( "neutral", real_ending_neutral_tones,  num_ending_neutral_tones_labeled )
    print()
    sum_r_end = len(real_ending_first_tones) + len(real_ending_third_tones) + len(real_ending_second_tones) + len(real_ending_fourth_tones) +len(real_ending_neutral_tones)
    sum_n_end = num_ending_first_tones_labeled + num_ending_third_tones_labeled + num_ending_second_tones_labeled + num_ending_fourth_tones_labeled + num_ending_neutral_tones_labeled
    print( "Congruence index percentage of ENDING tones: {pct}, r:{r}, n:{n}".format(pct=sum_r_end/sum_n_end,r=sum_r_end,n=sum_n_end))
    print()
    print("Invalid Results: {tone_list}".format(tone_list=invalid_results))
    print()
    print("**********{song_name}**********".format(song_name = song_name))

 
if __name__=="__main__": 
    main()
```
